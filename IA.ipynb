Ejemplo prueba:  
  
```python
from ultralytics import YOLO

# Cargar un modelo preentrenado (ejemplo: YOLOv8 small)
model = YOLO("yolov8s.pt")

# Hacer predicción sobre una radiografía de prueba
results = model.predict("radiografia_prueba.jpg", save=True, imgsz=640)

# Mostrar resultados
for r in results:
    print(r.boxes)  # Coordenadas y confianza de detecciones
```
